```python
import pandas as pd 
```

```python
url = 'https://en.wikipedia.org/wiki/Results_of_the_2014_Indian_general_election#Results_by_constituency'
```

```python
tables = pd.read_html(url)

tables
```

```python
type(tables)
```

```python
electionresults = tables[78]
electionresults

new= electionresults.drop(electionresults.columns[[1, 3, 7 , 8, 9, 10]], axis=1)
new.head(5)
new = new.rename(columns={0:'State',2:'Name',4:'Candidate',5:'Party',6:'Votes'})
new.head(5)
new = new.iloc[1:,:]
new.head()
```

```python

organizednew = new[["Name","State","Candidate","Party", "Votes"]]
organizednew
```
